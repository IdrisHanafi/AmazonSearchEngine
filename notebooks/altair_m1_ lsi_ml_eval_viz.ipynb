```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('summary_knn_m1_metrics_v4.csv')

df1 = df.melt(id_vars=['features', 'components', 'n_gram'], value_vars=['accuracy', 'macro_f1', 'micro_f1', 'weighted_f1'])
```

```python
df.rename(columns = {'features': 'text feature'}, inplace = True)
df1.rename(columns = {'variable': 'Metrics'}, inplace = True)
```

```python
chart_1 = alt.Chart(df[df['n_gram'] == 2], title = 'ngram = 2').mark_bar().encode(
    x='text feature:O',
    y='accuracy:Q',
    color='text feature:N',
    column='components:N'
)

chart_2 = alt.Chart(df[df['n_gram'] == 3], title = 'ngram = 3').mark_bar().encode(
    x='text feature:O',
    y='accuracy:Q',
    color='text feature:N',
    column='components:N'
)


chart_1 | chart_2
```

```python
chart_1 = alt.Chart(df[df['n_gram'] == 2], title = 'ngram = 2').mark_point().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='text features:N'
).properties(
    width=200,
    height=200
).repeat(
    row=['accuracy', 'macro_f1', 'micro_f1', 'weighted_f1'],
    column=['components']
)

chart_2 = alt.Chart(df[df['n_gram'] == 3], title = 'ngram = 3').mark_point().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='features:N'
).properties(
    width=200,
    height=200
).repeat(
    row=['accuracy', 'macro_f1', 'micro_f1', 'weighted_f1'],
    column=['components']
)

chart_1 
```

```python
range_ = ["#79A3CE", "#92568B"]

chart_1 = alt.Chart(df).mark_point().encode(
    x='components:Q',
    y='accuracy:Q',
    color=alt.Color('text feature:N',scale=alt.Scale(range=range_))
).properties(
    width=180,
    height=180
).facet(
    column='n_gram:N', title = "Accuracy of K-Nearest Neighbor Text Classifier"
).configure(background='#F2F2F2').configure_title(fontSize=15, offset=5, orient='top', anchor='middle')

chart_2 = alt.Chart(df).mark_point().encode(
    x='components:Q',
    y='macro_f1:Q',
    color='features:N'
).properties(
    width=180,
    height=180, title = "Accuracy of K-Nearest Neighbor Text Classifier"
).facet(
    column='n_gram:N'
)

chart_1
```
